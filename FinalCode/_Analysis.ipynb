```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

path="../Datasets/DIODE_FULL - Copia (2)"

df = pd.DataFrame(columns=['scene','scan','name','path_rgb','path_depth','path_mask'])
min_list=[]
mean_list=[]
max_list=[]
folder=os.listdir(path)
for scene in folder:
    scans=os.listdir(path+'/'+scene)
    for scan in scans:
        files=os.listdir(path+'/'+scene+'/'+scan)
        for file in files:
            if file[-4:]=='.png':
                path_base=path+'/'+scene+'/'+scan+'/'+file[:-4]
                last_ = path_base.rfind("_")
                if not path_base.endswith(('F', 'G', 'S', 'C', '_')):
                    print(path_base)
                    matrix=np.load(str(path_base+'_depth'+'.npy'))
                    min_list.append(matrix.min())
                    max_list.append(matrix.max())
                    mean_list.append(matrix.mean())
```

```python
cifre_decimal=1
min_round = [round(numero, cifre_decimal) for numero in min_list] 
max_round = [round(numero, cifre_decimal) for numero in max_list] 
mean_round = [round(numero, cifre_decimal) for numero in mean_list] 
```

```python
def print_distribution(lista):
    # Calcola il numero di occorrenze per ciascun valore
    valori, conteggi = np.unique(lista, return_counts=True)

    # Imposta il range dell'asse y
    plt.xlim(min(valori), max(valori))
    # Creazione del barplot
    plt.bar(valori, conteggi)

    # Etichettatura degli assi
    plt.xlabel('Valori')
    plt.ylabel('Numero di occorrenze')
    plt.title("Distribuzioni minimi valori per immagine")
    # Mostra il plot
    plt.show()

print_distribution(min_round)
print_distribution(max_round)
print_distribution(mean_round)
```
